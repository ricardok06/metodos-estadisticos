#Regresion simple


```python
import matplotlib.pyplot as plt
import math
x=[1.4,4.8,8.6,4.4,4.2,4.2,2.4,3,1.7,9.2]
y=[23.4,29.1,36.2,28.5,27.9,28.2,24.5,26.2,23.3,37]
q=0
#Operaciones-----------------------------------------------
n=len(x)
xy_s=0
x_s=0
y_s=0
x2_s=0
y2_s=0
sxy=0
for i in range(0,n):
  xy_s=xy_s+x[i]*y[i]

for i in range(0,n):
  x_s=x_s+x[i]

for i in range(0,n):
  y_s=y_s+y[i]

for i in range(0,n):
  y2_s=y2_s+y[i]**2

for i in range(0,n):
  x2_s=x2_s+x[i]**2

px=x_s/n
py=y_s/n

for i in range(0,n):
  sxy=sxy+(x[i]-px)*(y[i]-py)


a1=(n*xy_s-x_s*y_s)/(n*x2_s-x_s**2)
a0=(py)-(a1*px)

sxx=x2_s-1/n*x_s**2
syy=y2_s-1/n*y_s**2

r2=(sxy**2)/((sxx)*(syy))
r=math.sqrt(r2)

#Grafico------------------------------------------------------
x1 = [min(x)-1,max(x)+1]
y1 = [a1*i + a0 for i in x1]
plt.plot(x1, y1,color='#00875e')
plt.scatter(x, y, marker='o', color='#04394e',s=10)

#Ecuaciones----------------------------------------------------
p=a1*q+a0
eq = r'$y = ({})x+({})$'.format(round(a1, 4),round(a0, 4))
sxyeq= r'$sxy = {}$'.format(round(sxy,4))
sxxeq= r'$sxx = {}$'.format(round(sxx,4))
syyeq= r'$syy = {}$'.format(round(syy,4))
r2= r'$r^2 = {}$'.format(round(r2,4))
r= r'$r = {}$'.format(round(r,4))
yq= r'$y({}) = {}$'.format(q,round(p,4))


plt.figtext(1,0.8,eq, fontsize=20)
plt.figtext(1,0.7,sxyeq, fontsize=20)
plt.figtext(1,0.6,sxxeq, fontsize=20)
plt.figtext(1,0.5,syyeq, fontsize=20)
plt.figtext(1,0.4,r2, fontsize=20)
plt.figtext(1,0.3,r, fontsize=20)
plt.figtext(1,0.2,yq,size=20)

plt.show()


```

#Regresion polinomial


```python
import numpy as np
import math
import matplotlib.pyplot as plt
x=[35, 50, 65, 80, 95, 110]
y=[16, 26, 41, 62, 88, 119]
q=6

#Operaciones--------------------------------------------------
Sx=sum(x)
Sx_2=sum(x**2 for x in x)
Sx_3=sum(x**3 for x in x)
Sx_4=sum(x**4 for x in x)

Sy=sum(y for y in y)
Sx_y=sum(x*y for x, y in zip(x,y))
Sx_2_y=sum(x**2*y for x, y in zip(x,y))

coeficientes = np.array([[len(x),Sx, Sx_2],
                         [Sx, Sx_2, Sx_3],
                         [Sx_2, Sx_3,Sx_4]])
respuestas=np.array([Sy,Sx_y,Sx_2_y])
a=np.linalg.solve(coeficientes, respuestas)


sr=sum((y-a[0]-a[1]*x-a[2]*x**2)**2 for x,y in zip(x,y))
st=sum( (i-np.mean(y))**2 for i in y)
r_2=(st-sr)/st
r=math.sqrt(r_2)
syEx=math.sqrt((sr)/(len(x)-3))


#Grafico-----------------------------------------------------------
x1 = np.linspace(min(x)-1, max(x)+1, 100)
y1 = a[0]+a[1]*x1+a[2]*x1**2
plt.plot(x1, y1,color='#00875e')
plt.scatter(x, y, marker='o', color='#04394e',s=10)

#Ecuaciones---------------------------------------------------------

eq = r'$y = ({})+({})x+({})x^2$'.format(round(a[0], 4),round(a[1], 4),round(a[2], 4))
r_2eq= r'$r^2 = {}$'.format(round(r_2,4))
req= r'$r = {}$'.format(round(r,4))
syExeq = r'$S_{{y/x}} = {}$'.format(round(syEx, 4))
p=a[0]+a[1]*q+a[2]*q**2
yq= r'$y({}) = {}$'.format(q,round(p,4))

plt.figtext(1,0.8,eq,size=20)
plt.figtext(1,0.7,r_2eq,size=20)
plt.figtext(1,0.6,req,size=20)
plt.figtext(1,0.5,syExeq,size=20)
plt.figtext(1,0.4,yq,size=20)
plt.show()
```

#Regresion lineal multiple

```python
#import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x_1=[15, 16, 24, 13, 21, 16, 22, 18, 20, 16, 28, 27, 13, 22, 23 ]
x_2=[70, 65, 71, 64, 84, 86, 72, 84, 71, 75, 84, 79, 80, 76, 88 ]
y=[156, 157, 177, 145, 197, 184, 172, 187, 157, 169, 200, 193, 167, 170, 192 ]
q=[0,0]
#Operaciones------------------------------------------------
sumx_1=sum(x_1)
sumx_2=sum(x_2)
sumx_1a2=sum(i**2 for i in x_1)
sumx_1x_2=sum(i*j for i,j in zip(x_1,x_2))
sumx_2a2=sum(i**2 for i in x_2)
sumy=sum(y)
sumx_1y=sum(i*j for i,j in zip(x_1,y))
sumx_2y=sum(i*j for i,j in zip(x_2,y))

coeficientes = np.array([[len(x_1),sumx_1, sumx_2],
                         [sumx_1, sumx_1a2, sumx_1x_2],
                         [sumx_2, sumx_1x_2,sumx_2a2]])
respuestas=np.array([sumy,sumx_1y,sumx_2y])
a=np.linalg.solve(coeficientes, respuestas)

#Grafico-------------------------------------------------------
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x_1g = np.linspace(min(x_1)-1,max(x_1)+1, 3)
x_2g = np.linspace(min(x_2)-1,max(x_2)+1, 3)
X_1, X_2 = np.meshgrid(x_1g, x_2g)
Y = a[0]+a[1]*X_1+a[2]*X_2
surf = ax.plot_surface(X_1, X_2, Y, cmap='viridis', alpha=0.5)
ax.scatter(x_1, x_2, y, c='blue', marker='o')

#Ecuaciones-----------------------------------------------------------
eq = r'$y = ({})+({})x_1+({})x_2$'.format(round(a[0], 4),round(a[1], 4),round(a[2], 4))
eval=a[0]+a[1]*q[0]+a[2]*q[1]
evaleq= r'$y({},{}) = {}$'.format(q[0],q[1],round(eval,4))
plt.figtext(1,0.8,eq,size=20)
plt.figtext(1,0.7,evaleq,size=20)


plt.show()
```

#Polinomio de Lagrange


```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, simplify, latex
#funcion de el polinomio-----------------------------------
def lagrange(x, y):
    xsimbolico=symbols('x')
    L=0

    for i in range(len(x)):
        terma=1
        termb=1
        for j in range(len(x)):
            if i != j:
                terma*=(xsimbolico - x[j])
                termb*=(x[i] -x[j])

        L+= (terma)/(termb)*y[i]

    return L
#Datos-------------------------------------------------------
x=[-4, -1, 5, 9]
y=[-6, 0, 1, 8]
q=-4
polinomio=lagrange(x,y)


#Grafico-------------------------------------------------------
plt.scatter(x, y, marker='o', color='#04394e',s=10)
xgraf=np.linspace(min(x)-1, max(x)+1, 100)
ygraf=[polinomio.subs('x', i) for i in xgraf]
plt.plot(xgraf, ygraf,color='#00875e')

#ecuaciones---------------------------------------------------------
eq = r'$L(x)={}$'.format(latex(polinomio))
simplieq = r'$L(x)={}$'.format(latex(simplify(polinomio)))
eval=polinomio.subs('x',q)
evaleq = r'$L({})={}$'.format(q,eval)
plt.figtext(1,0.8,eq,size=20)
plt.figtext(1,0.7,simplieq,size=20)
plt.figtext(1,0.6,evaleq,size=20)
plt.show()
```

#Polinomio de Newton

```python
from sympy.core.numbers import sympify_complex
import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, simplify, latex

def matriz(x,y):
    n = len(x)
    matriz = np.zeros((n, n))
    matriz[:, 0] = y

    for j in range(1, n):
        for i in range(n - j):
            matriz[i][j] = (matriz[i + 1][j - 1] - matriz[i][j - 1]) / (x[i + j] - x[i])

    return matriz
def newton(matriz,x):
  n = len(x)
  xsimbolico = symbols('x')
  N = matriz[0][0]
  for i in range(1, n):
      term = matriz[0][i]
      for j in range(i):
          term *= (xsimbolico - x[j])
      N += term
  return simplify(N)


#Datos-------------------------------------------------------
x=[-4, -1, 5, 9]
y=[-6, 0, 1, 8]
q=9
matriz=matriz(x,y)
polinomio=newton(matriz,x)


#Grafico-------------------------------------------------------
plt.scatter(x, y, marker='o', color='#04394e',s=10)
xgraf=np.linspace(min(x)-1, max(x)+1, 100)
ygraf=[polinomio.subs('x', i) for i in xgraf]
plt.plot(xgraf, ygraf,color='#00875e')

#ecuaciones---------------------------------------------------------
eq = r'$N(x)={}$'.format(latex(polinomio))
eval=polinomio.subs('x',q)
evaleq = r'$N({})={}$'.format(q,eval)
plt.figtext(1,0.8,eq,size=20)
plt.figtext(1,0.7,evaleq,size=20)
plt.show()

```
